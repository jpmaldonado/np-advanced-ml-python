```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(random_state=1234)
```

```python
plt.scatter(X[:,0], X[:,1]);
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(X)
```

```python
labels = km.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=labels);
```

```python
km.inertia_
```

#### Elbow method

- Heuristic to calculate the number of clusters $k$ that is optimal.
- Plot different values of $k$ vs the corresponding inertia and pick the inflection point as a candidate for optimal $k$.



```python
Ks = [2,3,4,5,6,7,8]
inertias = [KMeans(n_clusters=k).fit(X).inertia_ for k in Ks]
```

```python
plt.plot(Ks, inertias)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia");
```

```python
km = KMeans(n_clusters=3)
labels = km.fit(X).predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
```
